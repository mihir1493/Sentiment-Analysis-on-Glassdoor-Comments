```python
import pandas as pd
```

```python
# Creating a dataframe to read the json file that includes the comments for each individual
df1 = pd.read_json('comcast_final.json')
```

```python
# Printing the dataframe
df1
```

```python
# Using pandas explode() function to split the list in cell elements into individual row element across each column category
# Use Pandas >= 0.25 version
df1 = df1.apply(pd.Series.explode).reset_index()
df1.head()
```

```python
# Splitting the date_position colmun into two individual columns called "date" & "job title"
df1[['date','job title']] = df1['date_position'].str.split("-", n=1, expand=True)
df1.head()
```

```python
# Checking the data type of each column using the .info() function
df1.info()
```

```python
# Converting the "date" column datatype from object to date format 
df1['date'] = pd.to_datetime(df1['date'])
```
