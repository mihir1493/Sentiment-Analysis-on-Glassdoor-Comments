```python
import pandas as pd
```

```python
# Creating a dataframe to read the json file that includes the comments for each individual
df = pd.read_json('glassdoor/items.json')
```

```python
# Printing the dataframe
df
```

```python
# Using pandas explode() function to split the list in cell elements into individual row element across each column category
# Use Pandas >= 0.25 version
df3 = df.apply(pd.Series.explode).reset_index()
df3
```

```python
# Splitting the date_position colmun into two individual columns called "date" & "job title"
df3[['date','job title']] = df3['date_position'].str.split("-", n=1, expand=True)
df3
```

```python
# Checking the data type of each column using the .info() function
df3.info()
```

```python
# Converting the "date" column datatype from object to date format 
df3['date'] = pd.to_datetime(df3['date'])
```

```python
df3.info()
```

```python
# Printing dataframe
df3
```

```python
# Finding the number of unique element in the dataframe column
df3.nunique()
```

```python
df3.reset_index(drop=True, inplace=True)
```

```python
# Droping "index" column from the dataframe
df3.drop(['index'], axis=1)
```

```python
# Exporting cleaned data as csv
df3.to_csv('cleaned_data.csv') 
```
