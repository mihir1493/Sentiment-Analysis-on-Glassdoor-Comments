```python
# Using pandas explode() function to split the list in cell elements into individual row element across each column category
# Use Pandas >= 0.25 version
df1 = df1.apply(pd.Series.explode).reset_index()
df2 = df2.apply(pd.Series.explode).reset_index()
df3 = df3.apply(pd.Series.explode).reset_index()

df1.head()
```
